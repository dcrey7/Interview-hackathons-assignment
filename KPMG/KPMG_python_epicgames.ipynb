# question 4
Getting friday and sunday critics recommend average

```python
df['date release']=pd.to_datetime(df['date release'])
```

```python
#formatting date 
import datetime
df['date release']=df['date release'].apply(lambda i:datetime.date.strftime(i, "%m/%d/%Y") )
```

```python
#getting day out of date 
import calendar
def findDay(date):
    born = datetime.datetime.strptime(date, '%m/%d/%Y').weekday()
    return (calendar.day_name[born])

df['day']=df['date release'].apply(lambda i: findDay(i))
```

```python
df.info()
```

```python
#cleaning critics recommend
df['Critics Recommend']=df['Critics Recommend'].apply(lambda i: i.replace('%',''))
df['Critics Recommend']=df['Critics Recommend'].apply(lambda i: int(i))
```

```python
df5=df[(df['day']=='Sunday') | (df['day']=='Friday')]
df5.groupby(['day'])['Critics Recommend'].mean()
```
